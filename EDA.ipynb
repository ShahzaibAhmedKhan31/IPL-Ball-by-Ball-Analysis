```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ballByball = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
matches = pd.read_csv("IPL Matches 2008-2020.csv")
```

```python
# percentage of missing values in each column.

matches.isnull().mean() * 100
```

 In matches dataset, there are 5 columns which are of not much use. They are
 Neutral_venue, method, result and eliminator. So, I will drop them.

```python
matches=matches.drop(["neutral_venue", "method", "result", "eliminator"], axis=1)
```

```python
#Checking for number of null values in each column

matches.isnull().sum()
```

```python
#Filling up the missing values in the city column
# Filling "city" column with "Mumbai" beacuse it is the most occuring value.

matches = matches.fillna({"city": matches["city"].mode()[0]})
matches
```

```python
matches.isnull().sum()
```

```python
# dropping those 4 rows in which player of the match was not there.

matches=matches[matches["player_of_match"].notna()]
matches
```

```python
matches.isnull().sum()
```

We are getting 13 rows in which result_margin is null. This simply means that those 13 matches were tied and went to the super over.

```python
ballByball_2 = ballByball[ballByball['inning'] == 1]
inning_1_bowling_team_counts = ballByball_2['bowling_team'].value_counts()
```

```python
inning_1_bowling_team_counts
```

```python
#filling the bowling_team null values with the name of the team that mostly end up at bowling side in the first innings.
#Which is Kalkata Knight Riders.
greatest_count_team = inning_1_bowling_team_counts.index[0]

ballByball['bowling_team'] = ballByball['bowling_team'].fillna(greatest_count_team)
```

```python
ballByball.isnull().sum()
```

```python

# filter the rows where either team1 or team2 is 'Chennai Super Kings' and the year is 2008
csk_matches = matches[(matches['team1'] == 'Chennai Super Kings') | (matches['team2'] == 'Chennai Super Kings')]

print(csk_matches.count())
```

The null values of dismissal_kind, player_dismissed,fielder and extras_type can be null as this data set is a ball by ball dataset, hence we will left them as it is.
